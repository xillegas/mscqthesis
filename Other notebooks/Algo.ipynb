```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
import numpy as np

# Crear un circuito de 8 qubits
n_qubits = 8
qc = QuantumCircuit(n_qubits)

# --- Paso 1: Inicialización con Hadamard ---
qc.h(range(n_qubits))

# --- Paso 2: Oráculo de Grover para |10000000> ---
# En notación de Qiskit, el estado se escribe como |q7 q6 ... q0>.
target_state = "10000000"

# Mapeamos |target_state> a |11111111> con compuertas X en los bits 0
for qubit, bit in enumerate(reversed(target_state)):
    if bit == "0":
        qc.x(qubit)

# Fase -1 solo para |11111111> usando un MCZ construido con H-MCX-H
qc.h(n_qubits - 1)
qc.mcx(list(range(n_qubits - 1)), n_qubits - 1)
qc.h(n_qubits - 1)

# Deshacemos el mapeo con X
for qubit, bit in enumerate(reversed(target_state)):
    if bit == "0":
        qc.x(qubit)

# --- Paso 3: Simulación para verificación ---
backend = Aer.get_backend("statevector_simulator")
job = backend.run(qc)
result = job.result()
statevector = result.get_statevector()

target_index = int(target_state, 2)
print("Vector de estado después del oráculo:")
print(statevector)
print(f"\nAmplitud del estado |{target_state}> (índice {target_index}):")
print(statevector[target_index])
```
